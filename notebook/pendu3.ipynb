Pour simplifier la conception du jeu, dans un premier temps on fixe le mot à chercher dans le programme, dans une variable nommée `mot`. Pour cela, rajouter dans le programme l'instruction suivante :

Il nous faut donc afficher six carrés, un pour chacune des lettres. Ce qui pourrait se faire par les six appels suivants :
1. `carre(-150, -250)` 
1. `carre(-100, -250)` 
1. `carre(-50, -250)` 
1. `carre(0, -250)` 
1. `carre(50, -250)`
1. `carre(100, -250)` 

On constate que dans ces appels, seul le paramètre `x` varie. La boucle va nous simplifier la tâche ; nous allons écrire un seul appel qui sera répété six fois :

````python
x = -150
for compteur in range(6):
    carre(x, -250)
    x += 50
````

Détaillons le fonctionnement de cette suite d'instructions :

- la variable `x` contient l'abscisse du coin inférieur gauche du carré 
- ligne 1 : `x` est initialisé à la valeur de la première abscisse -150
- ligne 4 : une fois un carré tracé, on calcule l'abscisse du suivant en ajoutant 50 à `x`
- la ligne 2 permet de répéter les instructions indentées 6 fois. Le nom `compteur` est choisi arbitrairement. Cette variable permet à Python de comptabilier le nombre de passages dans la boucle.
