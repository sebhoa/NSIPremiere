## <font color="darkblue"> &diams; Première étape : trier la liste</font>

La première étape est de trier la liste des objets :
* soit par valeur décroissante (on remplira le sac avec les objets de plus grande valeur d'abord)
* soit encore par ordre de poids croisssant (on remplira le sac avec les objets les plus légers d'abord)
* soit par valeur par unité de poids décroissant (on remplira le sac avec les objets ayant la plus grande valeur/poids d'abord)

Afin de trier la liste des objets, on utilisera la méthode <code>sort</code> de Python (*sort* signifie trier en anglais). 

```python
# Tri 
objets.sort()
```

```python
# On vérifie maintenant que la liste est bien triée
print(objets)
```

On constate que Python a bien trié notre liste mais par **ordre alphabétique**. En effet, faute d'avoir fourni à Python un critère de tri de nos objets, il a trié par le premier critère rencontré c'est à dire le nom.

On peut préciser un critère de tri pour notre liste d'objets grâce à l'option <code>key</code> de la méthode sort. Par exemple la valeur est le deuxième élément de notre liste d'objet, on commence par créer une fonction valeur qui prendre en argument un objet et renvoie sa valeur :

```python
def valeur(objet):
    # rappellons que objet est une liste [nom,valeur,poids]
    # la valeur est donc l'élément d'indice 1
    return objet[1]
```

On précise maintenant que c'est la valeur qui est le critère de tri :

```python
# Tri par ordre de valeur de la liste des objets
objets.sort(key = valeur)
# On affiche pour vérifier
print(objets)
```

On y est presque ! Les objets sont classés par ordre croissant de valeur alors que l'on voulait l'ordre décroissant. Pour inverser l'ordre il suffit de préciser <code>reverse=True</code> dans la méthode sort :

```python
# Tri par ordre décroissant de valeur
objets.sort(key=valeur,reverse=True)
print(objets)
```
