## <font color="darkblue"> &diams; Premire tape : trier la liste</font>

La premire tape est de trier la liste des objets :
* soit par valeur dcroissante (on remplira le sac avec les objets de plus grande valeur d'abord)
* soit encore par ordre de poids croisssant (on remplira le sac avec les objets les plus lgers d'abord)
* soit par valeur par unit de poids dcroissant (on remplira le sac avec les objets ayant la plus grande valeur/poids d'abord)

Afin de trier la liste des objets, on utilisera la mthode <code>sort</code> de Python (*sort* signifie trier en anglais). 

```python
# Tri 
objets.sort()
```

```python
# On vrifie maintenant que la liste est bien trie
print(objets)
```

On constate que Python a bien tri notre liste mais par **ordre alphabtique**. En effet, faute d'avoir fourni  Python un critre de tri de nos objets, il a tri par le premier critre rencontr c'est  dire le nom.

On peut prciser un critre de tri pour notre liste d'objets grce  l'option <code>key</code> de la mthode sort. Par exemple la valeur est le deuxime lment de notre liste d'objet, on commence par crer une fonction valeur qui prendre en argument un objet et renvoie sa valeur :

```python
def valeur(objet):
    # rappellons que objet est une liste [nom,valeur,poids]
    # la valeur est donc l'lment d'indice 1
    return objet[1]
```

On prcise maintenant que c'est la valeur qui est le critre de tri :

```python
# Tri par ordre de valeur de la liste des objets
objets.sort(key = valeur)
# On affiche pour vrifier
print(objets)
```

On y est presque ! Les objets sont classs par ordre croissant de valeur alors que l'on voulait l'ordre dcroissant. Pour inverser l'ordre il suffit de prciser <code>reverse=True</code> dans la mthode sort :

```python
# Tri par ordre dcroissant de valeur
objets.sort(key=valeur,reverse=True)
print(objets)
```
