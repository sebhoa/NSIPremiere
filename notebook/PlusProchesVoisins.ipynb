## <font color="darkblue"> &diams; A propose de voisins ... </font>
Remarquons que pour déterminer les voisins, il faut disposer d'une **distance** entre deux données. Cette notion mathématique de distance est difficile à formaliser et à définir de façon précise à notre niveau. On a utilisé dans l'exemple ci-dessus la **distance euclidienne** c'est à dire celle que l'on peut mesurer entre deux points à l'aide d'une règle. On rappelle que si les coordonnées de ces deux points sont $(x_1,y_1)$ et $(x_2,y_2)$ alors la distance euclidienne entre ces deux points est :
$$ d = \sqrt{(x_2-x_1)^2+(y_2-y_1)^2} $$
Cela a été vu dès la classe de seconde en mathématique. On admettra que cette distance se généralise si les "points" ont plus de 2 coordonnées.

Enfin, d'autres distances peuvent être définies (voir la dernière partie de ce notebook).

## <font color="darkblue"> &diams; Distance entre deux iris !</font>

Afin de déterminer les "voisins plus proches" d'un iris, il faut pouvoir mesurer la distance entre deux iris qui sont représentés par les quatre caractéristiques. On note $(x_1,y_1,z_1,t_1)$ les caractéristiques du premier iris et $(x_2,y_2,z_2,t_2)$ celles du second. On prendra comme distance entre ces eux iris :
$$ d = \sqrt{(x_2-x_1)^2+(y_2-y_1)^2+(z_2-z_1)^2+(t_2-t_1)^2} $$
