#  <span style="color:darkred;"> &#10070; Le module Turtle de Python</span>

L'objectif de cette série de notebook est de programmer entièrement ce petit jeu. Nous commencerons par la création des graphismes de base qui nous servirons pour le dessin de la potence et du bonhomme.

## <span style="color:darkblue;"> &diams; Un crayon et un papier virtuels</span>
Pour jouer, nous avons besoin d'une **feuille de papier** et d'un **crayon**. En Python, ce sera un peu la même chose. La feuille de papier est un écran (**screen** en anglais), et le crayon est une **tortue** (**turtle** en anglais). 

#### <span style="color:darkblue;">&#9672; <bold>Remarque : </bold></span>
Le nom **turtle** fait référence au langage Tortue [LOGO](https://fr.wikipedia.org/wiki/Logo_(langage)#Quelques_exemples) popularisé par [S.Papert](https://fr.wikipedia.org/wiki/Seymour_Papert) en 1963. 

---
Ce crayon et ce papier virtuel n'existent pas de base dans Python, il faut les **importer** depuis un module. C'est le rôle de l'instruction suivante :

### Importer le module

```python
import turtle
```

### Prendre un _crayon_ et une _feuille_

Avec `turtle` ça donne ceci :

```python
feuille = turtle.Screen()
crayon = turtle.Turtle()
```

Notez bien que les noms `crayon` (et `feuille`) sont de notre choix, nous aurions pu par exemple les appeler  `stylo` et `papier`.

Après exécution de ces instructions, une fenêtre nommée `Python Turtle Graphics` a dû s'ouvrir quelque part sur votre écran, c'est notre papier virtuel. Au centre de cet écran figure une flèche orientée à droite, c'est notre crayon virutel.

#### <span style="color:darkred;">&#9672; <bold>Attention</bold></span> : 
Cette fenêtre doit rester ouverte, vous pouvez par exemple  la positionner à côté de celle où s'affiche ce notebook.

<div class="alert alert-block alert-info">

## <span style="color:darkblue;"> &diams; Propriétés du crayon</span>

### Choix des couleurs

`turtle`  connait quelques noms de couleurs (mais il parle anglais) : `'black'` (noir), `'white'`(blanc), `'blue'`(bleu), `'lightgrey'`(gris clair), `'cyan'` (cyan) ... Et on peut donc choisir la couleur de sa feuille ainsi que celle de son crayon

```python
# bgcolor est l'abbréviation de background color (couleur de fond)
feuille.bgcolor('beige')
```

## <font color="darkblue"> &diams; Autres manipulations du crayon</font>

### Annuler la dernière action

La méthode `undo` permet d'annuler la dernière action faite (avec le crayon ou la feuille), donc pour supprimer les deux traits tracés précédemment :

```python
crayon.undo()
crayon.undo()
```

### Relever ou abaisser le crayon

Pour tracer notre premier trait (la base de la potence), il faut donc commencer par relever le crayon (de façon à se déplacer sans dessiner). Les instructions sont donc :
1. relever le crayon grâce à l'instruction `penup`
2. se déplacer au point de coordonnées (-200,-150) grâce à l'instruction `goto` (*aller* en anglais)
3. abaisser le crayon grâce à l'instruction `pendown`
3. aller au point de coordonnées (-100,-150)

```python
crayon.penup()
crayon.goto(-200,-150)
crayon.pendown()
crayon.goto(-100,-150)
```

Pour tracer le montant **vertical** de la potence :
1. relever le crayon
2. aller en (-150,-150)
4. abaisser le crayon
5. aller en (-150,200)

```python
crayon.penup()
crayon.goto(-150,-150)
crayon.pendown()
crayon.goto(-150,200)
```

## <span style="color:darkblue;"> &diams; Tracer des cercles </span>

La tête du pendu sera un cercle, tracé à l'aide de `circle` en donnant le rayon qui est de 25:

```python
crayon.circle(25)
```

Ce n'est pas le résultat attendu car le tracé du cercle dépend de l'**orientation de la tortue**, c'est à dire  l'angle qu'elle fait avec l'axe horizontal en tournant dans le sens inverse des aiguilles d'une montre. Comme illustré sur la figure suivante avec un angle de 30°



C'est ce qui se passe sur un rapporteur, comme en mathématiques.

Le centre du cercle est donc toujours situé à gauche de l'orientation de la tortue. Donc pour dessiner la tête du pendu en bas de la corde, il faut orienter la tortue vers la gauche (c'est à dire à 180°).

Pour modifier l'orientation de la tortue vers le haut, deux possibilités :
* Fixer son orientation grâce à `setheading`, ici il faut indiquer un  angle de 180°.
* La faire tourner par rapport à son orientation actuelle grâce à `left` (tourner à gauche) ou `right` (tourner à droite).

```python
crayon.undo()
crayon.setheading(180)
crayon.circle(25)
```

<div class="alert alert-block alert-info">
<b>&#9889; A retenir : </b> <br>
<ul>
    <li> L'orientation de la tortue est l'angle qu'elle forme avec l'axe horizontal</li>
    <li> Cette orientation est modifiable grâce aux instructions <code>setheading</code> (fixe l'orientation à un angle donné), ou <code>left</code> et <code>right</code> (tourne à gauche ou à droite par rapport à l'orientation actuelle).</li>
    <li> <code>circle</code> permet de tracer un cercle dont on donne le rayon. </li>
    <li> Le centre se trouve à gauche de l'orientation de la tortue.</li>
    </ul>
</div>
